```python

import psycopg2

# Connection parameters
host = "0.0.0.0"
database = "postgres"
user = "postgres"
password = "mysecretpassword"
port = "5432"  # default port, change if necessary

# Establish the connection
conn = psycopg2.connect(
    dbname=database,
    user=user,
    password=password,
    host=host,
    port=port
)

# Create a cursor object using the connection
cur = conn.cursor()

cur.execute("""
   SELECT "LEAID" FROM lea
""")

# Fetch all results
tables = cur.fetchall()

# Print table names
print("Tables in the 'public' schema:")
for table in tables:
    print(table[0])

cur.close()
conn.close()
```
