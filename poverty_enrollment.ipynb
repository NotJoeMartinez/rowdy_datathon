```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data/combined_data.csv')

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['LEA_ENR'], df['STUDENTS_BELOW_POVERTY'], color='blue')
plt.title('Correlation between Total Enrollment and Students Below Poverty line')
plt.xlabel('Total Enrollment (LEA_ENR)')
plt.ylabel('Students Below Poverty')
plt.grid(True)

# Adjusting axis limits to zoom in
plt.xlim(0, 60000)  # Adjust as necessary based on your data
plt.ylim(0, 4000)   # Adjust as necessary based on your data

plt.show()
```
